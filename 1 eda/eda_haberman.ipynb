```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
haber = pd.read_csv("haberman.csv")
```

```python
print(haber.shape)
print(haber.columns)
```

```python
haber["status"].value_counts()
```

```python
haber.plot(kind='scatter',x='age',y='year')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(haber,hue="status",size=4)\
    .map(plt.scatter,"age","year")\
    .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid")
sns.pairplot(haber,hue="status",size=3)
plt.show()
```

```python
haber1 = haber.loc[haber["status"] == 1]
haber2 = haber.loc[haber["status"] == 2]
plt.plot(haber1["age"],np.zeros_like(haber1["age"]),'o')
plt.plot(haber2["age"],np.zeros_like(haber2["age"]),'o')
plt.show()
```

```python
sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haber, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```
